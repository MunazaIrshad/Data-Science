**Objective:**
Store owner wants to create an annual sales report for 2022. 
So that, store owner can understand their customers and grow 
more sales in 2023.

Questions:
1) Compare the sale and orders using a single chart.
2) Which month got the highest sales and orders.
3) Who purchased more-men or women in 2022.
4) What are the different order status in 2022.
5) List top 10 states contributing to the sales.
6) Relation between age and gender based on number of orders.
7) Which channel is contributing to maximum sales.
8) Highest selling category?


```python
# Import Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load/Read Data

Sales_Data = pd.read_excel('Store Data Analysis.xlsx')
```

```python
# Show Data

Sales_Data.head()
```

```python
# Data Information

Sales_Data.info()
```

```python
# 1) Compare the sale and orders using a single chart.

sns.lineplot(Sales_Data, x ='Qty', y = 'Amount')
```

```python
sns.barplot(Sales_Data, x ='Qty', y = 'Amount')
```

```python
Sales_Data.columns
```

```python
# 2) Which month got the highest sales and order
# Highest number of Orders 

Sales_Data.groupby('Qty')['Month'].value_counts()
```

### **March Has the highest number of Orders: 2799**

```python
Sales_Data.groupby('Qty')['Month'].value_counts().plot(kind='bar')
```

```python
#Sales_Data.groupby('Qty')['Month'].value_counts().plot(kind='bar') 
sns.barplot(Sales_Data, x='Month', y='Qty')
```

```python
# Highest Sales

Sales_Data.groupby('Month')['Amount'].sum()
```

### **April has the Highest Sales: 1829263**

```python
Sales_Data.groupby('Month')['Amount'].sum().plot(kind='bar')
```

```python
sns.barplot(Sales_Data, x='Month', y='Amount')
```

```python
# 3) Who purchased more-men or women in 2022.
gen_purch = Sales_Data.groupby('Gender')['Qty'].sum()
gen_purch 
```

### **Womens purchased more : 21702 Orers**

```python
# More Sale

gen_purch = Sales_Data.groupby('Gender')['Amount'].sum()
gen_purch
```

### **More Sales by women: 13562773**

```python
# 4) What are the different order status in 2022.

Sales_Data['Status'].unique()
```

```python
Sales_Data['Status'].value_counts()
```

```python
sns.countplot(Sales_Data, x = 'Status')
```

```python
# 5) List top 10 states contributing to the sales.

Sales_Data.groupby('ship-state')['Amount'].sum().sort_values(ascending=False)
```

```python
Sales_Data.groupby('ship-state')['Amount'].sum().sort_values(ascending=False).head(10)
```

```python
Sales_Data.groupby('ship-state')['Amount'].sum().sort_values(ascending=False).head(10).plot(kind='bar')
```

```python
# 6) Relation between age and gender based on number of orders.

Sales_Data.groupby(['Gender', 'Age Group'])['Qty'].sum()
```

```python
Sales_Data.groupby(['Gender', 'Age Group'])['Qty'].count().plot(kind='bar', stacked=True)
```

```python
# 7) Which channel is contributing to maximum sales.

Sales_Data.groupby('Channel ')['Amount'].sum()
```

```python
Sales_Data.groupby('Channel ')['Amount'].sum().sort_values(ascending=False).plot(kind='bar')
```

### **Amazon Contributing to maximum sales: 7519933**

```python
# 8) Highest selling category?

Sales_Data.groupby('Category')['Amount'].sum().sort_values(ascending=False).plot(kind='bar')
```
