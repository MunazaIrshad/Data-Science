```python
# Import Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load Titanic dataset from seaborn
titanic_data = sns.load_dataset('titanic')
```

```python
# Scatter Plot

titanic_data.plot(kind = 'scatter', x = 'age', y = 'fare')
```

```python
titanic_data.plot(kind = 'scatter', x = 'age', y = 'fare', color = 'orange')
```

```python
# Change color with Hex color codes
titanic_data.plot(kind = 'scatter', x = 'age', y = 'fare', color = '#00ecfc')
```

```python
titanic_data.groupby('sex')['age'].mean()
```

```python
titanic_data.groupby('sex')['age'].mean().plot(kind = 'bar')
```

```python
# x and y Labels
titanic_data.groupby('sex')['fare'].mean().plot(kind = 'bar')
plt.xlabel('Average Fare')
plt.ylabel('Gender')
```

### Seaborn Plots

```python
# Bar Plot

sns.barplot(data = titanic_data, x = 'survived', y = 'fare')
```

```python
sns.barplot(titanic_data, x = 'sex', y = 'fare', hue = 'class')
```

```python
sns.boxplot(titanic_data, x = 'sex', y = 'age', hue = 'class')
```

```python
sns.scatterplot(titanic_data, x = 'age', y = 'fare', hue = 'class')
```

```python
sns.scatterplot(titanic_data, x = 'age', y = 'fare', hue = 'sex')
```
